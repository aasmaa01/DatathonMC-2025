```python
import cv2
import numpy as np
import torch

def preprocess_image(path, img_height=32, img_max_width=512):
    # Load in grayscale
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    
    # Original size
    h, w = img.shape

    # Scale the height to `img_height`
    scale = img_height / h
    new_w = int(w * scale)
    
    # Resize
    img = cv2.resize(img, (new_w, img_height))

    # Pad to fixed width (black background)
    if new_w < img_max_width:
        padded_img = np.zeros((img_height, img_max_width), dtype=np.uint8)
        padded_img[:, :new_w] = img
    else:
        # If image is too wide, crop it
        padded_img = img[:, :img_max_width]

    # Normalize to [0, 1]
    normalized = padded_img.astype(np.float32) / 255.0

    # Add channel dimension (1, H, W)
    tensor = torch.tensor(normalized).unsqueeze(0)  # shape: (1, 32, img_max_width)

    return tensor
```

```python
from torch.utils.data import Dataset
import torch

class OCRDataset(Dataset):
    def __init__(self, df, image_folder, char_to_idx, img_height=32, img_max_width=512):
        self.df = df
        self.image_folder = image_folder
        self.char_to_idx = char_to_idx
        self.img_height = img_height
        self.img_max_width = img_max_width

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_path = self.image_folder + '/' + self.df.iloc[idx]['image']
        label_text = self.df.iloc[idx]['text']

        image = preprocess_image(img_path, self.img_height, self.img_max_width)
        label = torch.tensor([self.char_to_idx[c] for c in label_text], dtype=torch.long)

        return image, label

```

```python
import torch.nn as nn
import torch.nn.functional as F

class CRNN(nn.Module):
    def __init__(self, num_classes):
        super(CRNN, self).__init__()
        
        # CNN backbone
        self.cnn = nn.Sequential(
            nn.Conv2d(1, 64, 3, padding=1),     # (B, 64, 32, 512)
            nn.ReLU(),
            nn.MaxPool2d((2, 2)),               # (B, 64, 16, 256)

            nn.Conv2d(64, 128, 3, padding=1),   # (B, 128, 16, 256)
            nn.ReLU(),
            nn.MaxPool2d((2, 2)),               # (B, 128, 8, 128)

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)),               # (B, 256, 4, 128)

            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d((2, 1)),               # (B, 512, 2, 128)
        )

        self.rnn = nn.LSTM(
            input_size=512 * 2,  # 2 is height after final pooling
            hidden_size=256,
            num_layers=2,
            bidirectional=True,
            batch_first=False
        )

        self.fc = nn.Linear(256 * 2, num_classes)  # 2 for bidirectional

    def forward(self, x):  # x: (B, 1, 32, 512)
        x = self.cnn(x)    # x: (B, 512, 2, W)
        b, c, h, w = x.size()
        x = x.permute(3, 0, 1, 2).contiguous()  # x: (W, B, C, H)
        x = x.view(w, b, c * h)                 # x: (W, B, C*H)

        x, _ = self.rnn(x)                      # x: (W, B, hidden*2)
        x = self.fc(x)                          # x: (W, B, num_classes)

        return x
```
