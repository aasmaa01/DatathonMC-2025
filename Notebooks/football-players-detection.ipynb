```python
# Import necessary libraries
import pandas as pd
import os

# Step 1: Load the test CSV file with the correct path
test_csv_path = '/kaggle/input/mc-datathon-2025-football-players-detection/images_test.csv'
df_test = pd.read_csv(test_csv_path)

# Step 2: Create a submission dataframe with dummy predictions
submission = []
for index, row in df_test.iterrows():
    image_id = row['image_id']  # Use the correct column name from the CSV
    # Dummy prediction: empty list or [0, 0, 0, 0] (no detection)
    prediction = [0, 0, 0, 0]  # [x_min, y_min, x_max, y_max]
    submission.append([image_id, prediction])

# Step 3: Convert to DataFrame with the correct column name
submission_df = pd.DataFrame(submission, columns=['image_id', 'prediction'])

# Step 4: Save the submission file
submission_df.to_csv('/kaggle/working/submission.csv', index=False)

print("Submission file created at /kaggle/working/submission.csv. Ready to submit!")
```
